```python
import tensorflow as tf
```

```python
def mean_absolute_scaled_error(y_true, y_pred): 
    mae = tf.reduce_mean(tf.abs(y_true-y_pred))
    mae_naive_no_season = tf.reduce_mean(tf.abs(y_true[1:]-y_pred[:-1]))
    return mae/mae_naive_no_season

def evaluate_preds(y_true, y_pred):
    y_true = tf.cast(y_true, dtype=tf.float32)
    y_pred = tf.cast(y_pred, dtype=tf.float32)
    
    mse = tf.keras.metrics.mean_squared_error(y_true, y_pred)
    mae = tf.keras.metrics.mean_absolute_error(y_true, y_pred)
    mape = tf.keras.metrics.mean_absolute_percentage_error(y_true, y_pred)
    rmse = tf.sqrt(mse)
    mase = mean_absolute_scaled_error(y_true, y_pred)
    if mse.ndim > 0:
        mse = tf.reduce_mean(mse)
        mae = tf.reduce_mean(mae)
        mape = tf.reduce_mean(mape)
        rmse = tf.reduce_mean(rmse)
        mase = tf.reduce_mean(mase)
    
    metrics = {
        "Mean Absolute Error": mae.numpy(), 
        "Mean Squared Error": mse.numpy(), 
        "Root Mean Squared Error": rmse.numpy(), 
        "Mean Absolute Percentage Error": mape.numpy(),
        "Mean Absolute Scaled Error": mase.numpy()
    }
    return metrics
```

## Dense model

```python
HORIZON = 1
WINDOW_SIZE = 7
```

```python
def get_labelled_window(x, horizon=HORIZON):
    return x[:, :-horizon], x[:, -horizon:]

def make_windows(x, window_size=WINDOW_SIZE, horizon=HORIZON):
    window_step = np.expand_dims(np.arange(window_size+horizon), axis=0)
    window_indexes = window_step + np.expand_dims(np.arange(len(x)-(window_size+horizon-1)), axis=0).T
    windowed_array = x[window_indexes]
    windows, labels = get_labelled_window(windowed_array, horizon=horizon)
    return windows, labels
```

```python
len(depths)
```

```python
depths
```

```python
full_windows, full_labels = make_windows(depths)
len(full_windows), len(full_labels)
```

```python
def make_train_test_splits(windows, horizons, test_size=0.2):
    split_size = int(len(windows) * (1-test_size))
    train_windows = windows[:split_size]
    train_horizons = horizons[:split_size]
    test_windows = windows[split_size:]
    test_horizons = horizons[split_size:]
    return train_windows, test_windows, train_horizons, test_horizons
```

```python
train_windows, test_windows, train_horizons, test_horizons = make_train_test_splits(full_windows, full_labels)
len(train_windows), len(test_windows), len(train_horizons), len(test_horizons)
```

```python
import os

def create_model_checkpoint(model_name, save_path="C:/Users/nrpra/Documents/vscode_py/model_expereiments"):
    return tf.keras.callbacks.ModelCheckpoint(filepath=os.path.join(save_path, model_name), 
                                             verbose=0, save_best_only=True, monitor="val_loss")
```

```python
from tensorflow.keras import layers

train_features = tf.data.Dataset.from_tensor_slices(train_windows)
train_target = tf.data.Dataset.from_tensor_slices(train_horizons)
train_dataset = tf.data.Dataset.zip((train_features, train_target))
train_dataset = train_dataset.batch(1024).prefetch(tf.data.experimental.AUTOTUNE)

test_features = tf.data.Dataset.from_tensor_slices(test_windows)
test_target = tf.data.Dataset.from_tensor_slices(test_horizons)
test_dataset = tf.data.Dataset.zip((test_features, test_target))
test_dataset = test_dataset.batch(1024).prefetch(tf.data.experimental.AUTOTUNE)

train_dataset, test_dataset
```

```python
model_dense = tf.keras.Sequential([
    layers.Dense(128, activation="relu"), 
    layers.Dense(HORIZON, activation="linear")
])

model_dense.compile(loss="mae", 
                   optimizer=tf.keras.optimizers.Adam())

model_dense.fit(train_dataset, epochs=100, verbose=1, validation_data=test_dataset, 
               callbacks=[create_model_checkpoint("model_dense")])
```

```python
model_dense = tf.keras.models.load_model("model_expereiments/model_dense/")
model_dense.evaluate(test_dataset)
```

```python
def make_preds(model, input_data):
    forecast = model.predict(input_data)
    return tf.squeeze(forecast)
```

```python
model_dense_preds = make_preds(model_dense, test_windows)
model_dense_preds[:10]
```

## LSTM model

```python
HORIZON = 1
WINDOW_SIZE = 7
```

```python
full_windows, full_labels = make_windows(depths, window_size=WINDOW_SIZE, horizon=HORIZON)
train_windows, test_windows, train_horizons, test_horizons = make_train_test_splits(full_windows, full_labels)
```

```python
from tensorflow.keras import layers

train_features = tf.data.Dataset.from_tensor_slices(train_windows)
train_target = tf.data.Dataset.from_tensor_slices(train_horizons)
train_dataset = tf.data.Dataset.zip((train_features, train_target))
train_dataset = train_dataset.batch(1024).prefetch(tf.data.experimental.AUTOTUNE)

test_features = tf.data.Dataset.from_tensor_slices(test_windows)
test_target = tf.data.Dataset.from_tensor_slices(test_horizons)
test_dataset = tf.data.Dataset.zip((test_features, test_target))
test_dataset = test_dataset.batch(1024).prefetch(tf.data.experimental.AUTOTUNE)

train_dataset, test_dataset
```

```python
tf.random.set_seed(42)

inputs = layers.Input(shape=(7))
x = layers.Lambda(lambda x: tf.expand_dims(x, axis=1))(inputs)
# x = layers.LSTM(128, return_sequences=True)(x)
x = layers.LSTM(128, activation="relu")(x)
# x = layers.Dense(32, activation="relu")(x)
outputs = layers.Dense(HORIZON)(x)
model_lstm = tf.keras.Model(inputs, outputs)

model_lstm.compile(loss="mae", 
               optimizer=tf.keras.optimizers.Adam())

model_lstm.summary()
```

```python
model_lstm.fit(train_dataset, epochs=100, verbose=1, validation_data=test_dataset, 
               callbacks=[create_model_checkpoint("model_lstm")])
```

```python
model_lstm = tf.keras.models.load_model("model_expereiments/model_lstm/")
model_lstm.evaluate(test_dataset)
```

```python
model_lstm_preds = make_preds(model_lstm, test_windows)
model_lstm_preds[:10]
```

```python
# timesteps into the future
FUTURE_TIMESTEPS = 28

def make_future_into_forecast(values, model, into_future, window_size=WINDOW_SIZE) -> list:
    future_forecast = []
    last_window = values[-WINDOW_SIZE:]
    for _ in range(FUTURE_TIMESTEPS):
        future_pred = model.predict(tf.expand_dims(last_window, axis=0))
        print(f"Predicting on: \n{last_window} -> Prediction: {tf.squeeze(future_pred).numpy()}\n")
        
        future_forecast.append(tf.squeeze(future_pred).numpy())
        last_window = np.append(last_window, future_pred)[-WINDOW_SIZE:]
    return future_forecast
```
